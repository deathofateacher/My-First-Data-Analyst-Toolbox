```python
# Genel resim için ana fonksiyon:
def check_df(dataframe, head):
    print("##################### Shape/Satır ve Sütun #####################")
    print(dataframe.shape)
    print("##################### Types/Veri Tipleri #####################")
    print(dataframe.dtypes)
    print("##################### Head/İlk 10 satır #####################")
    print(dataframe.head(head))
    print("##################### Tail/Son 10 satır #####################")
    print(dataframe.tail(head))
    print("##################### NA/Boş veri #####################")
    print(dataframe.isnull().sum())
    print("##################### Quantiles/Sayısal Kritikler #####################")
    print(dataframe.describe([0, 0.05, 0.50, 0.95, 0.99, 1]).T)
```

```python
##şimdi bu kategorik olarak farklı yönden keşfedeceğimiz veri setini daha kolay-sistemli bir şekilde inceleyelim.
def cat_smmry(dataframe, col_name):
    print(pd.DataFrame(
        {col_name: dataframe[col_name].value_counts(),
         "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)}))
    print("##########################################")
```

```python
#geldik en güzel kısma. bu kategorik özel kolonları görsel olarak inceleyelim.
def cat_sum_vis(dataframe, col_name):
    print(pd.DataFrame({
        col_name: dataframe[col_name].value_counts(),
        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)}))
    
    print("##########################################")
    
    sns.countplot(x=dataframe[col_name], data=dataframe)
    plt.show(block=True)
```

#### eğer sütunda boş veri var ise: (dataframe("sütunismi").isnull().sum()>0:)
#### eğer sütunun veri tipi sayısal ise: ( str(cf[col].dtypes) not in ["category", "object", "bool"]

```python
def celal(dataframe, col):
    if dataframe[col].isnull().sum()>0:
        if str(dataframe[col].dtypes) not in ["category", "object", "bool"]:
            dataframe[col] = dataframe[col].fillna(dataframe[col].median())
        else:
            dataframe[col] = dataframe[col].fillna(dataframe[col].mode()[0])
    else:
        print("Bu veri setinde eksik veri yoktur")
```
